# yang ini tanpa file csv
import matplotlib.pyplot as plt

```python
# ini adalah sumbu y dan x nya jika tanpa file csv kita perlu membuat x dan y nya
plt.plot([1,2,3,4,5,6,7],[6,9,6,9,8,7,0])

# dan garfik yang kita tulis disini tidak hanya bisa satu tapi lebih
# tapi yang diganti hanya yang sumbu y nya saja 
plt.plot([1,2,3,4,5,6,7],[0,9,8,7,6,4,3])
plt.plot([1,2,3,4,5,6,7],[1,3,5,7,1,7,0])
plt.plot([1,2,3,4,5,6,7],[9,3,7,9,5,7,5])
plt.plot([1,2,3,4,5,6,7],[4,6,3,2,7,8,4])


# ini adalah sumbu ,x, dam ,y, nya dan cara penulisan sytaxnya ini gunanya untuk mengetahui atau menjelaskan sumbu x dan y
plt.xlabel('sumbu x')
plt.ylabel('sumbu y')

# dan ini adalah penulisan judul  grafik kita
plt.title('ini contoh grafik yang tanpa csv')

# ini adalah legendnya biar kita gak kebingungan dengan memberikan keteranggannnya
plt.legend(['keterangan 1','keterangan 2','keterangan 3','keterangan 3','keterangan 4','keterangan 5'])

plt.show()
```
