```python
# untuk melakukan visialisasi data kita memerlukan dua modul
import matplotlib as plt # visualisasi data
import seaborn as sns # visualisasi data
#output dar visualisasi data akan di arahkan ke notebook
%matplotlib inline 
# seringkali di kenal sebgai magic coment di jupyter yang di peruntuk kan untuk mengalihkan visualisasi nya ke nooteboke yang sama
# artinya kalo kita melakukan visualisasi outputnya akan ke nooteboke ini
```

```python
# untuk menvisualissasikan korelasion matrix
# disini hasil nya eror karena nati di carik anjir capek
sns.heatmap(data = df.corr())
```

```python
# disini kita akan melakukan bar plot
# tapi kita harus tau dulu jumlah setiap spesies dalam data kita
df['age'].value_counts()
```

```python
# dan data di atas bisa divisualisasikan dengan menggunakan bar plot
# kita dengan menambakan plot.bar()
df['age'].value_counts().plot.bar()
plt.tight_layout()# ini gunanya agar visualisasinya maximum atau sesuai dengan layar
plt.show()
```

```python
# kita juga bisa melakukan bar plot degan seaborn 
sns.countplot(data=df,x ='age')#di sini lebih spesifik
plt.tight_layout()
```

```python
# karena yang bar sudah kita akan ke pie chart
import matplotlib.pyplot as plt
df['age'].value_counts().plot.pie(autopct =' %1.1f%%', labels = None, legend = True)# disini legend itu untuk keterangan yang ini dibawah  makanya di kasih nilai True kalau di kasih False legend nya ilang 
#label dikasih none agar tidak tabrakan
plt.tight_layout()# kenapa kalau pie chart gak membandingkan 2 nilai kolom dalam data kita karena gak cocok kalo bar cocok aja mungkkin kaya yang dibawah
```

```python
# //line plot atau graf baris disini agak sedikit kompleks karena kita akan //
# melakukan subplot dengan nrows dan ncols
import matplotlib.pyplot as plt

fig,ax = plt.subplots(nrows =4 , ncols =2 , figsize = (15,14))

ax[0, 0].plot(df['age'],df['mental_health_rating'],marker = 'o')
ax[0,0].set_title("umur")

ax[1, 0].plot(df['age'],df['study_hours_per_day'],marker = 'o')
ax[1,0].set_title("belajar perjam dalam sehari")

ax[2, 0].plot(df['age'],df['social_media_hours'],marker = 'o')
ax[2,0].set_title("bermain sosial media")

ax[3, 0].plot(df['age'],df['netflix_hours'],marker = 'o')
ax[3,0].set_title("nonton netflix")

ax[0,1].plot(df['age'],df['part_time_job'],marker = 'o')
ax[0,1].set_title("gak tau ini apa")

ax[1, 1].plot(df['age'],df['sleep_hours'],marker = 'o')
ax[1,1].set_title("sama")

ax[2, 1].plot(df['age'],df['diet_quality'],marker = 'o')
ax[2,1].set_title("qualitas diet")


# social_media_hours//	netflix_hours//	part_time_job//	attendance_percentage	sleep_hours	diet_quality	exercise_frequency	parental_education_level	internet_quality	mental_health_rating	extracurricular_participation	exam_score
```

```python
import matplotlib.pyplot as plt # yang inidari chatgpt

fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(12, 10))  # karena banyak grafik, pakai 3x3 grid

df['age'].plot.line(ax=ax[0][0])
ax[0][0].set_title("Umur")

df['study_hours_per_day'].plot.line(ax=ax[0][1])
ax[0][1].set_title("Jam Belajar")

df['social_media_hours'].plot.line(ax=ax[0][2])
ax[0][2].set_title("Jam Media Sosial")

df['netflix_hours'].plot.line(ax=ax[1][0])
ax[1][0].set_title("Jam Nonton Netflix")

df['attendance_percentage'].plot.line(ax=ax[1][1])
ax[1][1].set_title("Persentase Kehadiran")

df['exercise_frequency'].plot.line(ax=ax[1][2])
ax[1][2].set_title("Frekuensi Olahraga")

df['mental_health_rating'].plot.line(ax=ax[2][0])
ax[2][0].set_title("Skor Kesehatan Mental")

df['exam_score'].plot.line(ax=ax[2][1])
ax[2][1].set_title("Skor Ujian")

# Kosongkan panel terakhir jika tidak dipakai
fig.delaxes(ax[2][2])

plt.tight_layout()
plt.show()


```

```python
# ini kita menggunakan bar.plot bejar dari chatgpt kita bisa membandingkan dua kolom sekalugus
sns.barplot(data = df , x = 'gender', y = 'social_media_hours')
plt.xlabel('jenis kelamin')
plt.ylabel("lama memainkan sosial media")
plt.tight_layout()
plt.show()
```

```python
# ini kita menggunakan box.plot bejar dari chatgpt kita bisa membandingkan dua kolom sekalugus
sns.boxplot(data = df , x = 'age', y = 'social_media_hours')
plt.xlabel('jenis kelamin')
plt.ylabel("lama memainkan sosial media")#sleep_hours	social_media_hours
plt.tight_layout()
plt.show()
```
