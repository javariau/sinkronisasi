```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('student_habits_performance.csv')
df
```

```python
df['study_hours_per_day'] , df['gender'].value_counts().plot.bar()
plt.title('perbandingan antara laki-laki dan perempuan dalam belajar')
plt.xlabel('kelamin')
plt.ylabel('lama belajar')
plt.tight_layout()
plt.show()
```

```python
# disini kita akan mecoba menggunakan seaborn
sns.countplot(data = df , x = 'study_hours_per_day', hue = 'gender') #yang ini gak bisa kita tentukan sumbu y nya jadi sumbu y nya itu jumlah dari selurh data x
plt.tight_layout()
plt.show()
```

```python
df.plot()
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(nrows =3 , ncols =3 , figsize = (10,10))
df.age.plot.line(ax=ax[0][0],marker = 'o')
ax[0][0].set_title("umur")

df.study_hours_per_day.plot.line(ax=ax[0][1],marker = 'o')
ax[0][1].set_title("belajar perjam dalam sehari")

df.social_media_hours.plot.line(ax=ax[0][2],marker = 'o')
ax[0][2].set_title("bermain sosial media")

df.netflix_hours.plot.line(ax=ax[1][0],marker = 'o')
ax[1][0].set_title("nonton netflix")

df.attendance_percentage.plot.line(ax=ax[1][1],marker = 'o')
ax[1][1].set_title("gak tau ini apa")

df.exercise_frequency.plot.line(ax=ax[1][2],marker = 'o')
ax[1][2].set_title("ini juga")

df.exercise_frequency.plot.line(ax=ax[2][0],marker = 'o')
ax[2][0].set_title("sama")

df.mental_health_rating.plot.line(ax=ax[2][1],marker = 'o')
ax[2][1].set_title("mental healt")

df.exam_score.plot.line(ax=ax[2][2],marker = 'o')
ax[2][2].set_title("exam skor nya")
# yang kaya gini lebih mudah di baca
```

```python
# bisa aja gini tapi ini langsung semuanya jadi susah ngebacanya
df.plot(marker = 'o')
plt.tight_layout()# tapi data seperti ini susah di baca jika datanya banyak 
```

```python
# kita akan menggunakan pie chart
df['age'].value_counts().plot.pie(autopct = '%1.1f%%',label =None, legend =True)
plt.tight_layout()
```
